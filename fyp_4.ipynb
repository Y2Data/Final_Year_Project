```python
import numpy as np
from PIL import Image
from os import path
import matplotlib.pyplot as plt
import os
import random
from wordcloud import WordCloud, STOPWORDS
```

```python
gatsby_original = open("./Gatsby.txt", 'r', encoding='utf-8').read()
mask = np.array(Image.open("./car.jpg"))
```

```python
stopwords = list(STOPWORDS)

wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, margin=10, background_color='white',random_state=42).generate(gatsby_original)
```

```python
default_colours = wc.to_array()
plt.figure(figsize=(20,40))
plt.imshow(wc.recolor(), interpolation="bilinear")
```
